```python
import random
import matplotlib.pyplot as plt
```

```python
# Sierpinski triangle mde without using turtle
def triangle_plot(p):  # to make blank plot
    p_x = [x for (x,y) in p]
    p_y = [y for (x,y) in p]
    plt.title(f'Sierpinski triangle n = {iter_num}')
    plt.plot(p_x,p_y, '.', color='green')
    plt.show()

def triangle(iter_num):  # to make triangles
    v = [(0.0, 0.0), (0.5, 1.0), (1, 0.0)]
    p = []
    x,y = random.choice(v)
    for _ in range(iter_num):
        new_x,new_y = random.choice(v)
        x = (new_x + x) / 2.0
        y = (new_y + y) / 2.0
        p.append((x,y))
    triangle_plot(p)
iter_num = int(input())
triangle(iter_num)
# to make really beatigul plot, use big number, e.g. 9999
```

```python
# Sierpinski carpet plot
def carpet_plot(p):  # same function as previous
    p_x = [x for (x,y) in p]
    p_y = [y for (x,y) in p]
    plt.title(f'Sierpinski carpet n = {iter_num}')
    plt.plot(p_x,p_y, '.', color='blue')
    plt.show()

def carpet(iter_num):
    v = [(0.0, 0.0), (0.0, 0.5), (0.0, 1), (0.5, 1), (1, 1), (1, 0.5), (1, 0.0), (0.5, 0)]
    p = []
    x,y = random.choice(v)
    for _ in range(iter_num):
        new_x,new_y = random.choice(v)
        x = ((2*new_x) + x) / 3.0
        y = ((2*new_y) + y) / 3.0
        p.append((x,y))
    carpet_plot(p)

iter_num = int(input())

carpet(iter_num)
```

```python
input_text = input()
text = input_text.split()
new_text = []
for i in text:
    if len(i) > 3:
        old_word = list(i)
        first_letter = old_word[0]  # saving 1st and last letters
        last_letter = old_word[-1]
        old_word = old_word[1:-1]
        random.shuffle(old_word)  # shuffle
        old_word.append(last_letter)
        old_word.insert(0, first_letter)
        new_word = "".join(old_word)
        new_text.append(new_word)
    else:
        new_text.append(i)
print(" ".join(new_text))
```
